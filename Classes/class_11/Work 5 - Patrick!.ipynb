### HALLAMOS LA CANTIDAD TOTAL DE DESAYUNOS , DE MENÚS ECONOMICOS Y EL MONTO GASTADO EN AMBOS 

```python
# CON LA AYUDA DEL GROUPBY y LA AYUDA DEL SUM ; SUMAMOS LA VARIABLE "Concepto","cantidad","monto Total" , 
# COMO PODEMOS OBSERVAR SOLO EXISTEN 2 CONCEPTOS , EN LA CUAL SE HA SUMADO EL TOTAL DE CANTIDAD Y MONTO TOTAL , 
# DE TODOS A QUIENES ENCUENTRA COMO TALES VARIABLES , PERO RECORDEMOS QUE NO CUENTA A TODOS LAS VARIABLES , 
# DEBIDO A QUE NO LO RECONOCEN , SIENDO NaN

suma_conceptos = data.groupby( ['Concepto'] )[['cantidad', 'monto Total']].sum()
suma_conceptos
```
