## 1. Eliminar las columnas que crean necesaria

```python
la_libertad=base_libertad.drop(columns='Unnamed: 0' )
la_libertad.head()
```

## 2. Encontrar el tipo de datos

```python
la_libertad.dtypes
```

## 3. Encontrar la cantidad de habitantes

```python
lalibertad = la_libertad.filter(items = ['4_6_years_total','6_14_years_total'])
# lalibertad.groupby(['Region']).agg({'4_6_years_men':['sum'],'4_6_years_women':['sum'] }).sum(axis=1)
lalibe=lalibertad.sum(axis=0).sum(axis=0)
lalibe
```

## 4. Filtrar la cantidad de habitantes de 4 a 6 años en el distrito de trujillo


```python
trujillo=la_libertad.loc[la_libertad['District']== "TRUJILLO"]
trujillo
```

## 5. Localizar la region de mayor cantidad de habitantes tanto para personas de 4 a 6 años como de 6 a 14 años

```python
lalibe_4_6= la_libertad.filter(items = ['District','4_6_years_total'])
lalibe_6_14= la_libertad.filter(items = ['District','6_14_years_total'])
lalibe_4_6_aggr=lalibe_4_6.groupby(['District']).agg({'4_6_years_total':['sum']}).reset_index()
lalibe_4_6_aggr.columns=['District','4_6_years_total']
lalibe_4_6_aggr=lalibe_4_6_aggr.sort_values('4_6_years_total', ascending=False)
lalibe_4_6_aggr
```

```python
lalibe_6_14_aggr=lalibe_6_14.groupby(['District']).agg({'6_14_years_total':['sum']}).reset_index()
lalibe_6_14_aggr.columns=['District','6_14_years_total']
lalibe_6_14_aggr=lalibe_6_14_aggr.sort_values('6_14_years_total', ascending=False)
lalibe_6_14_aggr
```
