## ***Filter Data with Regular Expresions***

[link RE](https://docs.python.org/es/3/library/re.html)

**str.contains( pat, case = True, flags = 0, na = None, regex = True)**

**pat : str** <br>
* Character sequence or regular expression.

**case : bool, default True**
* If True, case sensitive.<br>

**flags : int, default 0(no flags)**<br>
* Flags to pass through to the re module, e.g.re.IGNORANCE<br>

**na : scarlar,optional**<br>
* Fill value for missing values. The default depends on dtype of the array. <br>

**regex : bool,default True**<br>
* If True, assumes the pat is a regular expression.<br>
* If False, treats the pat as a literal string.

## Special Characters 

**\d** : digits (0-9)<br>
**\D** : no digits (0-9)<br>
**\w** : character(a-z,A-Z,0-9,_ )<br>
**\W** : no character<br>
**\s** : white space (space, tab)<br>
**\S** :No white space (space, tab)<br>
**.**  : any character except newline<br>
**\** :  cancel special characters<br>
**^** : Start of a character string<br>
**$** : end of a string characters<br>

**( * )** : Makes the resulting RE match 0 or more repeats of the preceding RE<br>
**+**  : Makes the resulting RE match 1 or more repeats of the preceding RE
**?** : Makes the resulting RE match 0 or 1 repeats of the preceding RE
**( )** :  groups<br>
**[ ]** :  find characters in brackets<br>
**[a-z]** :  all lowercase letters<br>
**[ a-zA-Z ]** :  all lowercase and uppercase letters<br>
**[^a ]** : find any character except with a <br>
**[ a-zA-Z0-9 ]** : all numbers and letters included<br>
**|** :  conditional o<br>
**\b** : word limit<br>
**\B** : no word limit<br>
**\1** :  references<br>

### examples

```python
junin.loc[junin['District'].str.contains('ciudad', flags = re.I)]
```

```python
junin.loc[junin['Place'].str.contains('del', flags = re.I)]
```

```python
junin.loc[junin['District'].str.contains('yauli', flags = re.I)]
```

```python
junin.loc[junin['District'].str.contains('^hu', flags = re.I, na = False, regex = True)]
```

```python
junin.loc[junin['District'].str.contains('^huari', flags = re.I, na = False, regex = True)]
```

```python
junin.loc[junin['Place'].str.contains('ro$', flags = re.I, na = False, regex = True)]
```

```python
junin.loc[junin['Place'].str.contains('ca$', flags = re.I, na = False, regex = True)]
```

```python
junin.loc[junin['Place'].str.contains('am*', flags = re.I, na = False, regex = True)]
# match : a , m o am
```

```python
junin.loc[junin['Place'].str.contains('am+', flags = re.I, na = False, regex = True)]
# match : am (strict)
```

```python
junin.loc[junin['Place'].str.contains('am?', flags = re.I, na = False, regex = True)]
# match a or am
```

```python
texto = ''' Hola Mundo.
Me gusta Python!!!!!
Mi primero numero de la suerte es 987-654-321
Mi segundo numero de la suerte es 876-543-210
Mi tercer numero de la suerte es 765-432-100
Mi cuarto numero de la suerte es 123-456-123-123-456
'''
print(re.search(r'Mundo.$', texto, flags = re.M)) 
print(re.findall(r'[^\w\s]', texto))
```

```python
texto_1 = '''
13-04-2021
2021-13-04
2021-04-13
'''
print(re.findall(r'\d{2}-\d{2}-\d{4}', texto_1))
```

```python
texto_2 = '''
usuario antonioarturoaltmiarnaosd
abc
10
'''
print(re.findall(r'[a-z]{2}', texto_2))
```
